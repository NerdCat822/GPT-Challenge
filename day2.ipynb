```python
from langchain.prompts import FewShotPromptTemplate, PromptTemplate
from langchain.chains import LLMChain
from langchain.chat_models import ChatOpenAI
from langchain.callbacks import StreamingStdOutCallbackHandler

# ChatOpenAI 초기화
chat = ChatOpenAI(
    model_name="gpt-4o",
    streaming=True,
    callbacks=[StreamingStdOutCallbackHandler()],
    temperature=0.7
)

# 영화 정보 예시 데이터
movie_examples = [
    {
        "movie_name": "인셉션",
        "director": "크리스토퍼 놀란",
        "main_cast": "레오나르도 디카프리오, 마리옹 코티야르, 톰 하디, 조셉 고든-레빗",
        "budget": "1억 6천만 달러",
        "box_office": "8억 3천만 달러",
        "genre": "SF, 액션, 스릴러",
        "synopsis": "꿈 속에서 정보를 훔치는 도둑 코브가 마지막 임무로 '인셉션'을 수행하게 되는데, 이는 꿈 속에 아이디어를 심는 거의 불가능한 작업이다."
    },
    {
        "movie_name": "기생충",
        "director": "봉준호",
        "main_cast": "송강호, 이선균, 조여정, 최우식, 박소담",
        "budget": "1,140만 달러",
        "box_office": "2억 5,800만 달러",
        "genre": "드라마, 스릴러, 코미디",
        "synopsis": "반지하에 사는 기택 가족이 IT 기업 CEO 박사장의 집에 취업하면서 벌어지는 계급 갈등과 충돌을 그린 영화."
    }
]

# 영화 정보 출력 형식을 위한 예시 템플릿
example_template = """
영화명: {movie_name}
감독: {director}
주요 출연진: {main_cast}
예산: {budget}
흥행 수익: {box_office}
장르: {genre}
시놉시스: {synopsis}
"""

example_prompt = PromptTemplate(
    input_variables=["movie_name", "director", "main_cast", "budget", "box_office", "genre", "synopsis"],
    template=example_template
)
```

```python
# FewShotPromptTemplate 생성
few_shot_prompt = FewShotPromptTemplate(
    examples=movie_examples,
    example_prompt=example_prompt,
    prefix="당신은 영화 정보 전문가입니다. 주어진 영화에 대해 다음 형식으로 정보를 제공해주세요:",
    suffix="영화명: {movie_name}\n감독: {director}\n주요 출연진: {main_cast}\n예산: {budget}\n흥행 수익: {box_office}\n장르: {genre}\n시놉시스: {synopsis}",
    input_variables=["movie_name", "director", "main_cast", "budget", "box_office", "genre", "synopsis"]
)

# 영화 정보 체인 생성
movie_info_chain = LLMChain(
    llm=chat,
    prompt=few_shot_prompt,
    verbose=True
)
```

```python
# 영화 정보 체인 테스트
test_movie = {
    "movie_name": "어벤져스: 엔드게임",
    "director": "안소니 루소, 조 루소",
    "main_cast": "로버트 다우니 주니어, 크리스 에반스, 마크 러팔로, 크리스 헴스워스, 스칼렛 요한슨",
    "budget": "3억 5,600만 달러",
    "box_office": "27억 9,800만 달러",
    "genre": "액션, SF, 어드벤처",
    "synopsis": "타노스의 인피니티 건틀렛으로 인해 절반의 생명체가 사라진 후, 남은 어벤져스들이 과거로 돌아가 타임 스톤을 되찾아 사라진 동료들을 되살리려는 이야기."
}

result = movie_info_chain.run(**test_movie)
print("\n" + "="*50)
print("영화 정보 체인 테스트 결과:")
print("="*50)
print(result)
```

```python
# 더 간단한 사용법을 위한 함수 생성
def get_movie_info(movie_name):
    """
    영화 이름을 입력받아 영화 정보를 반환하는 함수
    """
    # 실제로는 여기서 영화 정보를 검색하거나 데이터베이스에서 가져와야 합니다
    # 예시를 위해 간단한 정보를 반환
    movie_data = {
        "movie_name": movie_name,
        "director": "감독 정보를 검색 중...",
        "main_cast": "출연진 정보를 검색 중...",
        "budget": "예산 정보를 검색 중...",
        "box_office": "흥행 수익 정보를 검색 중...",
        "genre": "장르 정보를 검색 중...",
        "synopsis": "시놉시스를 검색 중..."
    }
    
    return movie_info_chain.run(**movie_data)

# 사용 예시
print("영화 정보 조회 예시:")
print("="*30)
movie_info = get_movie_info("토이 스토리")
print(movie_info)
```

```python
# 간단한 체인 테스트
print("영화 정보 체인 테스트 시작...")
print("="*40)

# 테스트용 영화 데이터
test_movie_data = {
    "movie_name": "어벤져스: 엔드게임",
    "director": "안소니 루소, 조 루소", 
    "main_cast": "로버트 다우니 주니어, 크리스 에반스, 마크 러팔로, 크리스 헴스워스, 스칼렛 요한슨",
    "budget": "3억 5,600만 달러",
    "box_office": "27억 9,800만 달러", 
    "genre": "액션, SF, 어드벤처",
    "synopsis": "타노스의 인피니티 건틀렛으로 인해 절반의 생명체가 사라진 후, 남은 어벤져스들이 과거로 돌아가 타임 스톤을 되찾아 사라진 동료들을 되살리려는 이야기."
}

# 체인 실행
try:
    result = movie_info_chain.run(**test_movie_data)
    print("\n체인 실행 성공!")
    print("="*40)
    print(result)
except Exception as e:
    print(f"체인 실행 중 오류 발생: {e}")
```

```python
# 사용자 입력을 받아 영화 정보를 조회하는 함수
def search_movie_info(movie_name):
    """
    영화 이름만 입력받아 GPT가 영화 정보를 생성하도록 하는 함수
    """
    # GPT에게 영화 정보를 생성하도록 요청하는 프롬프트
    movie_search_prompt = f"""
    다음 영화에 대한 정보를 제공해주세요: {movie_name}
    
    다음 형식으로 답변해주세요:
    영화명: {movie_name}
    감독: [감독명]
    주요 출연진: [주요 배우들]
    예산: [제작비]
    흥행 수익: [전세계 박스오피스 수익]
    장르: [영화 장르]
    시놉시스: [영화 줄거리]
    """
    
    # 간단한 체인으로 실행
    simple_chain = LLMChain(llm=chat, prompt=PromptTemplate.from_template(movie_search_prompt))
    return simple_chain.run(movie_name=movie_name)

# 사용 예시
print("영화 검색 예시:")
print("="*30)
movie_result = search_movie_info("토이 스토리")
print(movie_result)
```
